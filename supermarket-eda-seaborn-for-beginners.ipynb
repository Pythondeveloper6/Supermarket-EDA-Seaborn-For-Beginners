### Seaborn :
- data visualisation library that helps in creating fancy data visualisations in Python
- specifically targets statistical data visualization
- as Michael Waskom says If matplotlib “tries to make easy things easy and hard things possible”, seaborn tries to make a well-defined set of hard things easy too.
- Seaborn provides a high-level interface to Matplotlib
- Seaborn works with different parameters
- Seaborn compliments and extends Matplotlib
- Seaborn has Built-in Seaborn Data Set

#### Seaborn Features :
- Using default themes that are aesthetically pleasing.
- Setting custom color palettes.
- Making attractive statistical plots.
- Easily and flexibly displaying distributions.
- Visualizing information from matrices and DataFrames.
- has Built-in Seaborn Data Set

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../input/supermarket-sales/supermarket_sales - Sheet1.csv')
```

```python
df.head()
```

```python
df.sample(10)
```

----
# Seaborn

```python
# Countplot
plt.figure(figsize=(12,8))
sns.countplot(x='Product line',data=df)
```

```python
# Countplot with hue
plt.figure(figsize=(12,8))
sns.countplot(x='Product line',hue='Gender',data=df)
```

-----

```python
plt.figure(figsize=(12,8))
sns.barplot(x='Product line',y='Total',hue='Gender',data=df,ci=None,estimator=sum)
```

------

```python
sns.boxplot(y='Total',data=df)
```

```python
sns.boxplot(x='Payment',y='Total',data=df)
```

```python
sns.boxplot(x='Payment',y='Total',data=df,showmeans=True)
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(x='Payment',y='Total',hue='Gender',data=df,showmeans=True)
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(x='Payment',y='Total',hue='Gender',data=df,showmeans=True,palette='PuOr_r')
```

```python
plt.figure(figsize=(16,6))
sns.boxplot(data=df,palette='PuOr_r')
```

-------

```python
sns.violinplot(x='Payment',y='Total',data=df)
```

```python
plt.figure(figsize=(12,8))
sns.violinplot(x='Payment',y='Total',hue='Gender',data=df)
```

```python
plt.figure(figsize=(12,8))
sns.violinplot(x='Payment',y='Total',hue='Gender',split=True,data=df)
```

-------

```python
sns.stripplot(y='Total',data=df)
```

```python
sns.stripplot(y='Total',x='Payment',data=df)
```

```python
sns.stripplot(y='Total',x='Payment',hue='Gender',data=df)
```

```python
sns.stripplot(y='Total',x='Payment',hue='Gender',data=df,dodge=True)
```

------

```python
sns.catplot(data=df,x='Payment',y='Total',hue='Gender')
```

```python
sns.catplot(data=df,x='Payment',y='Total',col='Gender')
```

```python
sns.catplot(data=df,x='Payment',y='Total',col='Gender',kind='box')
```

```python
sns.catplot(data=df,x='Payment',y='Total',col='Gender',kind='violin')
```
